```python
enData=pd.read_csv('/content/train.csv')
enData.head()
```

```python
from transformers import pipeline

def sentiment_analysis(tokenizer, model, dataset):

    sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
    predictions=sentiment_pipeline(dataset)

    return predictions

```

```python
from sklearn.utils import shuffle
trainData=shuffle(enData)[:100]
```

```python
results=sentiment_analysis(tokenizer,model,trainData['tweet'].tolist())
```

```python
results
```

```python
preds=[1 if d['label']=="LABEL_1" else 0 for d in results]
```

```python
print(classification_report(trainData['label'],preds))
```

seems that the model needs to be pre trained
