```python
import transformers
from transformers import pipeline
```

```python
ner=pipeline("ner",aggregation_strategy="simple",device=0)
```

```python
ner("Elon Musk founded Telsa")
```

```python
example_sentences = [
    "Apple Inc. is headquartered in Cupertino, California.",
    "In 1969, Neil Armstrong and Buzz Aldrin walked on the moon.",
    "The Eiffel Tower is a famous landmark in Paris, France.",
    "Microsoft Corporation was founded by Bill Gates and Paul Allen."
]
```

```python
preds=ner(example_sentences)
```
