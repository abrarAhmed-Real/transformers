```python
import transformers
from transformers import pipeline
import pandas  as pd
data=pd.read_csv('/content/bbc-text.csv')
data.head()
```

```python
input_text=data['text'].values[0]
```

```python
smm=pipeline("summarization")
```

```python
result=smm(input_text)
```
